```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
with open('Transacoes.txt', "r") as f:
    transactions = [line.strip().split(",") for line in f.readlines()]

transactions
```

```python
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df
```

```python
freq_item = apriori(df, min_support=0.5, use_colnames=True)
print(freq_item)
```

```python
rules = association_rules(freq_item, metric="confidence", min_threshold=0.5)
print(rules)
```

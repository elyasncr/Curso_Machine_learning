```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn import datasets
```

```python
iris = datasets.load_iris()  #Carregando o dataste

previsores = iris.data  #chamando os previsores do dataset

#gerando as classes do dataset
classe = iris.target


#mostrando os dados do dataset
previsores
```

```python
# Colocando os atributos do dataset no mesmo tipo de classe

sc = StandardScaler()
#transformando os dados
previsores = sc.fit_transform(previsores)

#Mostrando os dados após a transformação
previsores
```

No passo acima colocamos todos os dados na mesma escala, com o intuito de trazer todos com o mesmo padrã

```python
# São os atributos padronizados

previsores
```

```python
#Treinando as variáveis --> X_treinamento e y_treinamento
#Testando as variáveis --> X_teste e y_teste

X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe, test_size=0.3, random_state=123)
```

```python
#criando o modelo de treinamento
floresta = RandomForestClassifier(n_estimators=100, random_state=1234)
floresta.fit(X_treinamento, y_treinamento)

#modelo de teste
previsao = floresta.predict(X_teste)
taxa_acerto = accuracy_score(y_teste, previsao)

#vendo a taxa de acerto do teste
taxa_acerto
```

Acima vemos uma taxa de acerto de 93% calculando os dados após o treinamento e o tratamento dos dados.

```python
# Comparando a previsão e a taxa de acerto com o cálculo do PCA
pca = PCA(n_components=3)

# chamando um novo previsores.
previsores = pca.fit_transform(previsores)


# Resultado do PCA
previsores
```

o resultado do PCA são medidas artificialmente criadas através do modelo original.
Se repararmos bem, o modelo original conta com 4 colunas e neste há 3.
A partir disso, é possível fazer um novo treinamento e teste em cima dos novos previsores criados através do PCA.


```python
#Treinando as variáveis --> X_treinamento e y_treinamento
#Testando as variáveis --> X_teste e y_teste

X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe, test_size=0.3, random_state=123)
```

```python
# Novo treinamento em cima do resultado do PCA


#criando o modelo de treinamento PCA
floresta = RandomForestClassifier(n_estimators=100, random_state=1234)
floresta.fit(X_treinamento, y_treinamento)

#modelo de teste PCA
previsao = floresta.predict(X_teste)
taxa_acerto = accuracy_score(y_teste, previsao)

#vendo a taxa de acerto do teste do PCA
taxa_acerto
```

Acima vemos o resultado dos testes com os dados tratados pelo PCA.
Neste caso temos uma taxa de acerto de 95%, sendo maior do que o método tradicional.


## Considerações finais

dependendo de como for utilizado e qual será a tarefa necessária, para prever a taxa de acerto podemos utilizar duas maneiras.

1- método tradicional, tratando os dados manualmente podemos ter um X% de acerto
2- com o método do PCA, os dados são tratados artificialmente e em seguida ficam prontos para uso podendo trazer um X% maior de acerto.

ambos os métodos são bons, então fica a critério e necessidade do usuário qual usar.

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
dataset = pd.read_csv('credit_simple.csv', sep=';')
dataset.shape
```

```python
dataset.head()
```

```python
y = dataset['CLASSE']
X = dataset.iloc[:,:-1]
```

```python
X
```

```python
X.isnull().sum()
```

```python
mediana = X['SALDO_ATUAL'].median()
mediana
```

```python
X['SALDO_ATUAL'].fillna(mediana, inplace=True)
X.isnull().sum()
```

```python
agrupado = X.groupby(['ESTADOCIVIL']).size()
agrupado
```

```python
X['ESTADOCIVIL'].fillna('maculino solteiro', inplace=True)
X.isnull().sum()
```

```python
desv = X['SALDO_ATUAL'].std()
desv
```

```python
X.loc[X['SALDO_ATUAL']>=2*desv, 'SALDO_ATUAL']
```

```python
mediana_atual = X['SALDO_ATUAL'].median()
mediana_atual
```

```python
X.loc[X['SALDO_ATUAL']>=2*desv, 'SALDO_ATUAL'] = mediana_atual
X.loc[X['SALDO_ATUAL']>=2*desv]
```

```python
agroup = X.groupby(['PROPOSITO']).size()
agroup
```

```python
X.loc[X['PROPOSITO'] == 'Eletrodomésticos', 'PROPOSITO'] = 'outros'
X.loc[X['PROPOSITO'] == 'qualificação', 'PROPOSITO'] = 'outros'
agroup = X.groupby(['PROPOSITO']).size()
agroup
```

```python
X['DATA']
```

```python
X['DATA'] = pd.to_datetime(X['DATA'], format='%d/%m/%Y')
```

```python
X['DATA']
```

```python
X['ANO'] = X['DATA'].dt.year
X['MES'] = X['DATA'].dt.month
X['DIASEMANA'] = X['DATA'].dt.day_name()
```

```python
X['DIASEMANA']
```

```python
X['ESTADOCIVIL'].unique()
```

```python
X['PROPOSITO'].unique()
```

```python
X['DIASEMANA'].unique()
```

```python
label1 = LabelEncoder()
X['ESTADOCIVIL'] = label1.fit_transform(X['ESTADOCIVIL'])
X['PROPOSITO'] = label1.fit_transform(X['PROPOSITO'])
X['DIASEMANA'] = label1.fit_transform(X['DIASEMANA'])
```

```python
X.head()
```

```python
outros = X['OUTROSPLANOSPGTO'].unique()
outros
```

```python
z = pd.get_dummies(X['OUTROSPLANOSPGTO'], prefix = 'OUTROS')
```

```python
z
```

```python
X
```

```python
sc = StandardScaler()
m = sc.fit_transform(X.iloc[:,0:3])
m
```

```python
X = pd.concat([X,z,pd.DataFrame(m,columns=(['SALDO_ATUAL_N', 'RESIDENCIADESDE_N', 'IDADE_N']))], axis=1)
```
